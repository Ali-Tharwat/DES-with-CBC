#Bitwise Operations :

```python
def ASCIItoHex(ascii):
	hexa = ""
	for i in range(len(ascii)):
		ch = ascii[i]
		in1 = ord(ch)
		curr = hex(in1).lstrip("0x").rstrip("L")
		hexa += curr
	return hexa

def HextoBin(hexa):
    return bin(int(hexa, 16))[2:].zfill(len(hexa) * 4)

def BintoHex(bina):
    return hex(int(bina, 2))[2:].zfill(len(bina) // 4)

def HexToASCII(hexa):
    ascii = ""
    for i in range(0, len(hexa), 2):
        part = hexa[i : i + 2]
        ch = chr(int(part, 16))
        ascii += ch
    return ascii

def padding(binary):
    #We used Zero-Padding for its simplicity
    padding_length = 64 - (len(binary) % 64)
    padded_binary = binary + '0' * padding_length
    return padded_binary
```

# Round Function Methods :

```python
def XOR(a, b):
    result = ""
    for i in range(len(a)): result += "1" if a[i] != b[i] else "0"
    return result

def apply_sbox(s_box_num, bits):
    row = int(bits[0] + bits[5], 2)
    col = int(bits[1:5], 2)
    val = S_BOXES[s_box_num][row][col]
    return format(val, '04b')

def permute(binary, table):
    #Useful function for permutations :
    # key expansion , initial permutation , p-box and final permutation
    result = ""
    for pos in table:
        result += binary[pos - 1]  # -1 because positions in the table are 1-indexed
    return result
```

# DES Encryption for Round :

```python
def des_round(left, right, round_key, round_num):

    #Perform one round of DES encryption then return new left and right halves and shows each step's output

    print(f"\nRound {round_num + 1}:")
    print(f"Input: L{round_num} = {left}, R{round_num} = {right}")

    # Step 1: Expansion - Expand 32-bit right half to 48 bits
    expanded_right = permute(right, E)
    print(f"Expansion: E(R{round_num}) = {expanded_right}")

    # Step 2: Key mixing - XOR expanded right half with round key
    key_mixed = XOR(expanded_right, round_key)
    print(f"Key mixing: E(R{round_num}) ⊕ K{round_num + 1} = {key_mixed}")

    # Step 3: S-Box Substitution - Convert 48 bits to 32 bits
    substituted = ""
    for i in range(8):
        # Extract 6 bits for each S-Box
        six_bits = key_mixed[i*6:(i+1)*6]
        # Apply S-Box and get 4 bits
        four_bits = apply_sbox(i, six_bits)
        substituted += four_bits
    print(f"S-Box Substitution: S(E(R{round_num}) ⊕ K{round_num + 1}) = {substituted}")

    # Step 4: Permutation - Permute the 32-bit output from S-Boxes
    permuted = permute(substituted, P)
    print(f"Permutation: P(S(E(R{round_num}) ⊕ K{round_num + 1})) = {permuted}")

    # Step 5: XOR with left half
    new_right = XOR(left, permuted)
    print(f"XOR with left half: L{round_num} ⊕ f(R{round_num}, K{round_num + 1}) = {new_right}")

    # New left half is the old right half
    new_left = right
    print(f"Output: L{round_num + 1} = {new_left}, R{round_num + 1} = {new_right}")

    return new_left, new_right
```

# DES Encryption for Block :

```python
def des_encrypt_block(plainblock, round_keys):
   #BLOCK = 16 ROUNDS
    print("\n--- DES Block Encryption ---")
    print(f"Input 64-bit block: {plainblock}")

    # Step 1: Initial Permutation
    ip_output = permute(plainblock, IP)
    print(f"Initial Permutation (IP): {ip_output}")

    # Split the permuted block into left and right halves
    left = ip_output[:32]
    right = ip_output[32:]
    print(f"After splitting: L0 = {left}, R0 = {right}")

    # 16 rounds of DES
    for i in range(16): left, right = des_round(left, right, round_keys[i], i)

    # Swap left and right halves after the 16 rounds
    left, right = right, left
    print(f"\nAfter 16 rounds (with swap): L16 = {left}, R16 = {right}")

    # Combine the halves
    combined = left + right

    # Final Permutation
    cipherblock = permute(combined, FP)
    print(f"Final Permutation (FP): {cipherblock}")

    return cipherblock
```

# DES Encryption for CBC Mode :

```python
def des_cbc_encrypt(plaintext, iv, round_keys):

    # Convert plaintext to binary
    hexa = ASCIItoHex(plaintext)
    binary = HextoBin(hexa)
    print(f"\n--- DES CBC Encryption ---")
    print(f"Plaintext: '{plaintext}'")
    print(f"Binary representation: {binary}")

    # Pad the binary to a multiple of 64 bits
    padded_binary = padding(binary)
    print(f"Padded binary: {padded_binary}")
    print(f"Binary length: {len(binary)} bits")
    print(f"Padding length: {len(padded_binary) - len(binary)} bits")

    # Split into 64-bit blocks
    blocks = [padded_binary[i:i+64] for i in range(0, len(padded_binary), 64)]
    print(f"Number of 64-bit blocks: {len(blocks)}")

    # Initialize previous block with IV
    prev_block = iv
    ciphertext_blocks = []

    # Process each block
    for i, block in enumerate(blocks):
        print(f"\n===== Block {i+1} =====")
        print(f"Block {i+1} plaintext: {block}")

        # XOR with previous ciphertext block (or IV for first block)
        xor_result = XOR(block, prev_block)
        print(f"After XOR with {'IV' if i==0 else f'previous ciphertext block'}: {xor_result}")

        # Encrypt the XORed block
        cipherblock = des_encrypt_block(xor_result, round_keys)

        # Save ciphertext for next block's XOR
        prev_block = cipherblock
        ciphertext_blocks.append(cipherblock)

        # Display the final ciphertext for this block
        print(f"Block {i+1} ciphertext: {cipherblock}")
        print(f"Block {i+1} ciphertext (hex): {BintoHex(cipherblock)}")
        print(f"Block {i+1} ciphertext (ASCII): {HexToASCII(BintoHex(cipherblock))}")

    # Combine all ciphertext blocks
    full_ciphertext = ''.join(ciphertext_blocks)
    print("\n=== Final Results ===")
    print(f"Complete Ciphertext (binary): {full_ciphertext}")
    print(f"Complete Ciphertext (hex): {BintoHex(full_ciphertext)}")
    print(f"Complete Ciphertext (ASCII): {HexToASCII(BintoHex(full_ciphertext))}")

    return HexToASCII(BintoHex(full_ciphertext))
```
